## Exploratory Data Analysis (EDA)

### Explore ...

```python
products_df = products_df.merge(product_category_name_translation_df, on='product_category_name', how='left')

products_df
```

Menambahkan translation nama produk dengan melakukan merge dengan product_category_name_translation_df.

```python
orders_itemsorder_df = pd.merge(
    left=orders_df,
    right=order_items_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)

orders_itemsorder_df.head()
```

Menggabungkan dataset orders dan order_items untuk memperoleh data yang lebih lengkap dan terhubung

```python
customer_orders_itemsorder_df = pd.merge(
    left=customers_df,
    right=orders_itemsorder_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)

customer_orders_itemsorder_df
```

Selanjutnya, dataset digabungkan dengan dataset customer

```python
    city_customer = customer_orders_itemsorder_df.groupby("customer_city")
    
    sales_df = city_customer.price.sum().sort_values(ascending=False).reset_index()
    sales_df.rename(columns={
        "customer_city": "city"
    }, inplace=True)
    
    sales_df.head(8)
```

Didapatkan kota-kota dengan pembelian terbanyak, yaitu sao paulo (1,859,556.83), rio de janeiro (955,573.97), belo horizonte (346,039.04), dst.

```python
count_customer = city_customer.customer_id.count().sort_values(ascending=False).reset_index()
count_customer.rename(columns={
    "customer_city": "city",
    "customer_id": "customer_count"
}, inplace=True) 

count_customer.head(8)
```

Jumlah customer terbanyak ditemukan di kota sao paulo, rio de janeiro, belo horizonte, brasilia, curitiba, campinas, porto alegre, dan salvador

```python
products_customer_orders_itemsorder_df = pd.merge(
    left=products_df,
    right=customer_orders_itemsorder_df,
    how="left",
    left_on="product_id",
    right_on="product_id"
)

products_customer_orders_itemsorder_df
```

```python
products_customer_orders_itemsorder_df['order_purchase_month'] = pd.to_datetime(products_customer_orders_itemsorder_df['order_purchase_timestamp']).dt.to_period('M')
products_customer_orders_itemsorder_df
```

```python
sales = products_customer_orders_itemsorder_df.groupby("product_category_name_english").price.sum().sort_values(ascending=False).reset_index()
sales.rename(columns={
    "product_category_name_english": "Product Category",
    "price": "Price"
}, inplace=True)

sales
```

```python
monthly_trend = products_customer_orders_itemsorder_df.groupby("order_purchase_month").price.sum().reset_index()

monthly_trend.rename(columns={
    "order_purchase_month": "month",
    "price": "revenue"
}, inplace=True)

monthly_trend["month"] = monthly_trend["month"].dt.to_timestamp().dt.strftime('%Y-%m')
latest_year = monthly_trend.tail(13)
latest_year
```

```python
products_customer_orders_itemsorder_df.to_csv("all_df.csv", index=False)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:
#### Kota apa yang sering membeli produk dari e-commerce ini?

```python
top_city_consumer = sales_df.head(8)
colors_ = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
sns.barplot(
    x="price", 
    y="city",
    data=top_city_consumer.sort_values(by="price", ascending=False),
    palette=colors_,
    hue=top_city_consumer['city']
)

plt.title("Number of Sales by City", loc="center", fontsize=15)
plt.ylabel("City")
plt.xlabel("Sales ($)")
plt.tick_params(axis='y', labelsize=12)

plt.savefig("Numbers_of_Sales_by_City.png", format='png', dpi=300)

plt.show()
```

Secara signifikan, Sao Paulo dan Rio de Janeiro melakukan pembelian yang lebih banyak dibandingkan kota-kota lain. Tepatnya senilai \\$1,859,556.83, disusul oleh Rio de Janeiro sebesar \\$955,573.97. Hal ini berarti e-commerce ini memiliki basis pembeli yang besar di Sao Paulo dan Rio de Janeiro.

### Pertanyaan 3 : Bagaimana tren penjualan dalam setahun terakhir?


```python
top_product = sales.head(8)
colors_ = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
sns.barplot(
    x="Price", 
    y="Product Category",
    data=top_product,
    palette=colors_,
    hue=top_product['Product Category']
)

plt.title("Most Purchased Products", loc="center", fontsize=15)
plt.ylabel("Product Category")
plt.xlabel("Price ($)")
plt.tick_params(axis='y', labelsize=12)

plt.savefig("Most_Purchased_Products.png", format='png', dpi=300)

plt.show()
```

Secara berurutan, penjualan produk tidak berbeda jauh satu sama lain. Produk dengan kategori health beauty menyumbang penjualan yang paling banyak, yaitu sebesar \\$1,233,131.72. Kemudian disusul oleh Watches Gifts sebesar \\$1,165,898.98 dan bed_bath_table sebesar $1,023,434.76. 

```python
plt.figure(figsize=(10, 5))
plt.plot(
    latest_year["month"],
    latest_year["revenue"],
    marker='o',
    linewidth=2,
    color="#72BCD4"
)
plt.title("Total Revenue 2017-08 to 2017-08", loc="center", fontsize=20)
plt.xticks(fontsize=10, rotation=90)
plt.yticks(fontsize=10)

plt.savefig("revenue_in_a_year.png", format='png', dpi=300)


plt.show()
```

Secara garis besar, penjualan mengalami peningkatan selama setahun terakhir. Penjualan bulan Agustus 2018 sebesar \\$836,377.59, meningkat dibandingkan setahun sebelumnya yang bernilai \\$549,731.25. Namun, terjadi penurunan penjualan selama 3 bulan terakhir dengan nilai yang kurang lebih sama, yaitu 
\\$862,874.47 pada bulan Juli, \\$853,204.84 pada bulan Juni.
